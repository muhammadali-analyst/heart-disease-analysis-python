# 18. Compare Resting Blood Pressure As Per Sex Column

```python
data['sex'] = data['sex'].replace({1: 'Male', 0: 'Female'})

sns.histplot(data=data, x='trestbps', hue='sex', bins=20, kde=True)
plt.title('Resting Blood Pressure Distribution by Gender')
plt.xlabel('Resting BP (trestbps)')
plt.ylabel('Count')
plt.show()
```

# 19. Plot Continuous Variables

```python
data.columns
```

```python
categ_val=[]
cont_val=[]

for column in data.columns:
    if data[column].nunique() <=10:
        categ_val.append(column)
    else:
        cont_val.append(column)
```

```python
categ_val
```

```python
cont_val
```

```python
data.hist(cont_val,figsize=(15,6))
plt.tight_layout()
plt.show()
```
